```python
# импорты
import pandas as pd
import xlrd
from pandas_profiling.profile_report import ProfileReport
```

```python
# чтение данных в список датафреймов
dfs = []
rb = xlrd.open_workbook('data_last.xlsx')
for item in rb.sheets():
    if item.name[:2].lower() == 'uz':
        df = pd.read_excel('data_last.xlsx', item.name)
        df['well_id'] = item.name
        dfs.append(df)
```

```python
# колонки с данными для анализы + индексы
cols = ['well_id', 'TVD', 'GR', 'WNKT', 'DTP', 'DS_diff']
```

```python
# удаление неиспользуемых колонок из датафреймов
for df in dfs:
    df_cols = list(df.columns)
    df.drop(columns=list(set(df_cols) - set(cols)), inplace=True)
```

```python
df_full = pd.concat(dfs)
df_full.to_pickle('data.pkl')
df_full.set_index(['well_id', 'TVD'])
```

```python
report = ProfileReport(df_full)
report
```
